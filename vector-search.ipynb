```python
# Appended the ISBN as identifier, so we can split of that ISBN from the front of the description and use that as the thing to filter
# Filtering
# Fazer isso via text mach seria muito lento e bem confuso
# Convert the Series to a single string with each element on a new line.
output_string = books["tagged_description"].str.cat(sep="\n")

# Write the resulting string to the file.
with open("tagged_description.txt", "w", encoding="utf-8") as f:
    f.write(output_string)
```

```python
# raw_documents = TextLoader("tagged_description.txt").load()
# text_splitter = CharacterTextSplitter(chunk_size=0,chunk_overlap=0, separator="\n") # se chunk_size >1 tem grande chance not split on a new line
# documents =  text_splitter.split_documents(raw_documents)

# Load the documents from the file
raw_documents = TextLoader("tagged_description.txt").load()

# Initialize the text splitter with a large chunk_size
# This allows the splitter to use the newline separator to create new documents,
# without breaking individual lines into smaller chunks.
text_splitter = CharacterTextSplitter(
    chunk_size=1000,  # Use a large enough number, like 1000 or more
    chunk_overlap=0,
    separator="\n"
)

# Split the documents
documents = text_splitter.split_documents(raw_documents)
```
